```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')
import os
from pandas.plotting import scatter_matrix
import pandas as pd


import seaborn as sns
```

```python
WORKING_DIR = "D:\\proj\\toolbox\\"
INPUT_FILE  = "\\sample_data\\rtb100k.csv"
print(WORKING_DIR+INPUT_FILE)
os.getcwd()
```

```python
# dependent variable
dv = 'convert'

# Independent Variables
iv = ["6","7","10","11","12","13","14"]
```

```python
df = pd.read_csv(WORKING_DIR + INPUT_FILE, delimiter=';')
print(df.info())
df.head()
```

```python
stats  = df.describe().transpose()
stats
```

```python
for i, col in enumerate(stats.index):
  plt.figure(i)
  plt.title(col)
  sns.distplot(df[col])
```

```python
plt.close()
```

```python


scatter_matrix(df[ iv+[dv]], alpha=0.5, diagonal='kde',figsize=(16, 16)) 
plt.suptitle('scatter-matrix')

plt.show()
```
